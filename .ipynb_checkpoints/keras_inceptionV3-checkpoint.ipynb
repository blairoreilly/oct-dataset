```python
# load keras modules
import keras
from keras import applications
from keras.preprocessing.image import ImageDataGenerator
from keras import optimizers
from keras.models import Sequential
from keras.layers import Dropout, Flatten, Dense, GlobalAveragePooling2D
from keras.models import Model
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
# specify image information
# image dims
img_width, img_height = 256, 256

# define directories and parameters
train_data_dir = 'data/train'
test_data_dir = 'data/test'
validation_data_dir = 'data/validation'
train_samples = 83484
validation_samples = 968
epochs = 20
batch_size = 32
```

Build the Model

build the pretrained Inception V3 network 

compile the model using categorical cross-entropy loss function

Add new layers on top of the original model. 

add a global average pooling layer, a fully connected layer, dropout, and softmax activation with Adam optimizer at default

```python
# build Inception V3 network with pretrained weights from ImageNet
base_model = applications.InceptionV3(weights='imagenet', 
                                      include_top=False, # remove top fully connected layers
                                      input_shape=(img_width, img_height, 3))

# set base model to non trainable
for layer in base_model.layers:
    layer.trainable=False
```

```python
# build classifier model for top of convolutional model
# global average pooling layer and fully connected layer
# apply dropout and softmax activation for multiclass

model_top = Sequential()
model_top.add(GlobalAveragePooling2D(
    input_shape=base_model.output_shape[1:],
    data_format=None)),
model_top.add(Dense(256, activation='relu'))
model_top.add(Dropout(0.5))
model_top.add(Dense(4, activation='softmax'))
model = Model(inputs=base_model.input, 
              outputs=model_top(base_model.output))

# for custom metrics
import keras.backend as K

def mean_pred(y_true, y_pred):
    return K.mean(y_pred)


# compile model using Adam optimizer, categorical cross entropy loss
# use low learning rate for transfer learning
model.compile(optimizer=Adam(lr=0.0001, beta_1=0.9, beta_2=0.999, 
                             epsilon=1e-08, decay=0.0),
              loss='categorical_crossentropy',
              metrics=['accuracy', mean_pred])
```

```python
# define train and val generator and fit model
# image dims
img_width, img_height = 256, 256

# define directories and parameters
train_data_dir = 'data/train'
test_data_dir = 'data/test'
validation_data_dir = 'data/validation'
nb_train_samples = 83484
nb_validation_samples = 968
epochs = 20
batch_size = 32


# fix data format issues
if K.image_data_format() == 'channels_first':
    input_shape = (3, img_width, img_height)
else:
    input_shape = (img_width, img_height, 3)
    

from keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(rescale = 1./ 255, 
                             width_shift_range=0.2,
                             height_shift_range=0.2,
                             shear_range=0.2,
                             zoom_range=0.2,
                             horizontal_flip=True,
                             rotation_range=20)


#x , y = generator.next()

# data generator for test set
test_datagen = ImageDataGenerator(rescale = 1./255)

# generator for reading train data from folder
train_generator = datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_width, img_height),
    color_mode = 'rgb',
    batch_size = batch_size,
    class_mode = 'categorical',
    shuffle = True)

# generator for reading validation data from folder
# set shuffle=False for generator_predict/ generator.reset()
validation_generator = datagen.flow_from_directory(
    validation_data_dir,
    target_size=(img_width, img_height),
    color_mode = 'rgb',
    batch_size = batch_size,
    class_mode = 'categorical',
    shuffle = False)

# generator for reading test data from folder
test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=(img_width, img_height),
    batch_size = 1,
    class_mode = None,
    shuffle = False)

# fit pretrained Inception V3 model using data generator
history = model.fit_generator(
    train_generator,
    steps_per_epoch=train_samples // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=validation_samples // batch_size)
```

```python
# save model architecture
model.save_weights('inception.h5')
with open('incept_architecture.json', 'w') as f:
    f.write(model.to_json())
```
